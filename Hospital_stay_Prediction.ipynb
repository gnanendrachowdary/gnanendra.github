#  CHECKING OUTIERS

```python
for col in numeric:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
print(outliers)
```

```python
for col in outliers:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot for {col}')
    plt.show()
```

* We identified that outliers are there in columns and we need to detect the outliers by capping means Replace extreme values with upper or lower thresholds  

 # AFTER FILLING CHECKING OUTLIERS

```python
def cap_outliers(df, columns):
    for col in columns:
        lower_limit = df[col].quantile(0.05)  # 5th percentile
        upper_limit = df[col].quantile(0.95)  # 95th percentile
        df[col] = np.where(df[col] < lower_limit, lower_limit, df[col])
        df[col] = np.where(df[col] > upper_limit, upper_limit, df[col])
    return df

# List of numerical columns to cap
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Apply the capping function
df = cap_outliers(df, numerical_cols)
```

* Now checking whether outliers are filled or not

```python
for col in outliers:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x=col)
    plt.title(f'Boxplot for {col}')
    plt.show()
```

We are done with outliers now moving on to Model Building

# MODEL BUILDING

```python
# Splitting the data
x=df.drop("Stay",axis=1)
y=df["Stay"]

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

# K_NEAREST NEIGHBORS

```python
# Initialize and train the model
knn = KNeighborsClassifier(n_neighbors=5)  
knn.fit(x_train, y_train)

# Make predictions
y_pred_knn = knn.predict(x_test)

# Evaluate the model
print("KNN Accuracy:",round(knn.score(x_train,y_train)*100,2))
```

# HYPER PARAMETER TUNING FOR KNN

```python
param_grid_knn = {
    'n_neighbors': [5, 7],
    'weights': ['uniform']
}

grid_knn = GridSearchCV(estimator=knn, param_grid=param_grid_knn, cv=5, scoring='accuracy')
grid_knn.fit(x_train, y_train)

print("Best Parameters for KNN:", grid_knn.best_params_)
```

```python
best_knn=grid_knn.best_estimator_
y_pred_best_knn=best_knn.predict(x_test)

# Evaluate the model
acc_knn = round(grid_knn.score(x_train, y_train) * 100, 2)
print("Accuracy:",acc_knn)
```

After tuning the model also, it is not performing well , so we move on to another model

# DECISION TREE CLASSIFIER

```python
# Initialize and train the model
dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)

# Make predictions
y_pred_dt = dt.predict(x_test)

#Evaluation
acc_decision_tree = round(dt.score(x_train, y_train) * 100, 3)
print("Accuracy:",acc_decision_tree)
```

# TUNING DECISION TREE MODEL

```python
# Define the parameter grid
param_grid_dt = {
    'max_depth': [5, 10, 15,None],  # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  
    'min_samples_leaf': [1, 2, 4],    
    'criterion': ['gini', 'entropy'] 
}
```

```python
# Initialize Decision Tree classifier
dt = DecisionTreeClassifier(random_state=42)

grid_dt = GridSearchCV(estimator=dt, param_grid=param_grid_dt, cv=5, scoring='accuracy')
grid_dt.fit(x_train, y_train)

# Print the best parameters
print("Best Parameters for Decision Tree:", grid_dt.best_params_)
```

```python
best_dt = grid_dt.best_estimator_

# Make predictions on the test set
y_pred_best_dt = best_dt.predict(x_test)

print("Tuned Decision Tree Accuracy:",round(best_dt.score(x_train,y_train)*100,2))
```

# RANDOM FOREST

```python
# Initialize and train the model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)

# Make predictions
y_pred_rf = rf.predict(x_test)

# Evaluate the model
print("Random Forest Accuracy:",round(rf.score(x_train,y_train)*100,3))
```
